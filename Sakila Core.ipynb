```python
import pandas as pd

import pymysql
pymysql.install_as_MySQLdb()
```

```python
from sqlalchemy import create_engine
```

```python
connection = "mysql+pymysql://root:root@localhost/sakila"
```

```python
engine = create_engine(connection)
```

```python
engine
```

```python
q = """SELECT * FROM country;"""
pd.read_sql(q, engine)
```

```python

sql = """
select first_name, last_name, email, address_id, city_id
from customer join city
where city_id = 312;
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select title, name as genre, description, release_year, rating, special_features
from category join film
where name = 'Comedy';
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select last_name, title, release_year
from actor join film
where last_name = 'Lollobrigida';
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select first_name, last_name, title
from film join actor
where title = 'Bingo Talented';
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
select customer_id, amount
from payment
where amount >
(select 2* avg(amount))
"""

pd.read_sql_query(sql, engine)
```

```python
select first_name, last_name, amount
from payment join customer
where amount = max(amount)
```
