```python
import pandas as pd

import pymysql
pymysql.install_as_MySQLdb()
```

```python
from sqlalchemy import create_engine
```

```python
connection = "mysql+pymysql://root:root@localhost/sakila"
```

```python
engine2 = create_engine(connection)
```

```python
engine2
```

```python
q = """SELECT * FROM country;"""
pd.read_sql(q, engine2)
```

```python
#1
sql = """
select first_name, last_name, email, address_id, city_id
from customer join city on customer.customer_id = customer.customer_id
where city_id = 312;
"""

pd.read_sql_query(sql, engine2)
```

```python
#2
sql = """
select title, name as genre, description, release_year, rating, special_features
from category join film on film.title = film.title
where name = 'Comedy';
"""

pd.read_sql_query(sql, engine2)
```

```python
#3
sql = """
select last_name, title, release_year
from actor join film on film.title =film.title
where last_name = 'Lollobrigida';
"""

pd.read_sql_query(sql, engine2)
```

```python
#4
sql = """
select first_name, last_name, title
from film join actor on film.title = film.title
where title = 'Bingo Talented';
"""

pd.read_sql_query(sql, engine2)
```

```python
sql = """
select customer_id, amount
from payment
where payment.amount >
(select(2*avg(payment.amount))from payment)
"""

pd.read_sql_query(sql, engine2)
```

```python
sql = """
select customer.first_name, customer.last_name, count(payment.payment_id) AS num_payments, customer.customer_id
from customer join payment on customer.customer_id = payment.customer_id 
group by customer.customer_id
order by num_payments DESC
"""
pd.read_sql_query(sql, engine2)
```
